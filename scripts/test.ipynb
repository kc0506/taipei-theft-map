```python
import json
from pathlib import Path

import pandas as pd
import requests

DATA_PATH =Path(   "../data/car" )

df: pd.DataFrame|None = None

for fp in DATA_PATH.glob("*.json"):
    new_df=pd.DataFrame( json.loads( fp.read_bytes() ) )
    if df is None:
        df =new_df
    else:
        # df = df.
        df=pd.concat(( df,new_df ))


# df.drop([0,1],axis=1)
df=df.drop(['_id', '_importdate'], axis=1)
df=df    .set_index('編號')
df.to_csv(DATA_PATH/"car_data.csv")
df.iloc[:100]
```
